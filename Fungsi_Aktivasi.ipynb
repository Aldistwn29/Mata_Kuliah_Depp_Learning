```python
# Library
import keras
import numpy as np
import matplotlib.pyplot as plt
```

# **Fungsi Linear**

```python
# Model
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[2])])

# compile
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
class LossHistory(keras.callbacks.Callback):
  def on_train_begin(self, logs={}):
      self.losses = []
  def on_batch_end(self, batch, logs={}):
      self.losses.append(logs.get('loss'))
```

```python
# set data
xs = np.array([[1, 1], [1, 0], [0, 1], [0, 0]], dtype= int)
ys = np.array([1, 1, 1, 0], dtype= int)
```

```python
xs
```

```python
ys
```

```python
# tampilkan arsitektul awal
model.summary()
```

```python
history = LossHistory()
model.fit(xs, ys, epochs=500, callbacks=[history])
```

```python
loss_values = history.losses

plot1 = plt.figure(1)
plt.plot(loss_values)
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.show()
```

# **Fungsi Tanh**


```python
# Model2
model2 = keras.Sequential([keras.layers.Dense(units=1, input_shape=[2], activation= keras.activations.tanh)])

# model2 compile
model2.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# training 2
history2 = LossHistory()
model2.fit(xs, ys, epochs=500, callbacks=[history2])
```

```python
plot2 = plt.figure(2)

loss_values2 = history2.losses
plt.plot(loss_values2)
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.show()
```
